### Feature engineering

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
mm_scaler = MinMaxScaler()
ss_scaler = StandardScaler()
df_mm_scaled = pd.DataFrame(mm_scaler.fit_transform(df), columns=df.columns)
df_ss_scaled = pd.DataFrame(ss_scaler.fit_transform(df), columns=df.columns)
```

### Data preparation for modeling

```python
from sklearn.model_selection import train_test_split
x_mm_scaled = df_mm_scaled.drop('Chance of Admit ', axis=1)
y_mm_scaled = df_mm_scaled['Chance of Admit ']
x_ss_scaled = df_ss_scaled.drop('Chance of Admit ', axis=1)
y_ss_scaled = df_ss_scaled['Chance of Admit ']
```

```python
X_train_mm, X_test_mm, y_train_mm, y_test_mm = train_test_split(x_mm_scaled, y_mm_scaled, test_size=0.3, random_state=1)
X_train_ss, X_test_ss, y_train_ss, y_test_ss = train_test_split(x_ss_scaled, y_ss_scaled, test_size=0.3, random_state=1)
```

## 3. Model building

### Build the Linear Regression model and comment on the model statistics

```python
from sklearn.linear_model import LinearRegression
l_mm = LinearRegression()
l_ss = LinearRegression()
l_mm.fit(X_train_mm, y_train_mm)
l_ss.fit(X_train_ss, y_train_ss)
```

### Display model coefficients with column names

```python
#For data set which was scaled using Min Max:
for i, j in enumerate(X_train_mm.columns):
    print('Column \'', j, '\' has coefficient ', l_mm.coef_[i], sep='')
print('Intercept is', l_mm.intercept_)
```

```python
#For data set which was scaled using Standard Scaler:
for i, j in enumerate(X_train_ss.columns):
    print('Column \'', j, '\' has coefficient ', l_ss.coef_[i], sep='')
print('Intercept is', l_ss.intercept_)
```

### Try out Ridge and Lasso regression

```python
from sklearn.linear_model import Ridge, Lasso
rate_list = [0.01, 0.1, 1, 5, 10]
ridge_coef_mm = []
lasso_coef_mm = []
ridge_intercept_mm = []
lasso_intercept_mm = []
ridge_coef_ss = []
lasso_coef_ss = []
ridge_intercept_ss = []
lasso_intercept_ss = []
for i in rate_list:
    ridge_mm = Ridge(alpha=i)
    lasso_mm = Ridge(alpha=i)
    ridge_ss = Ridge(alpha=i)
    lasso_ss = Ridge(alpha=i)
    ridge_mm.fit(X_train_mm, y_train_mm)
    lasso_mm.fit(X_train_mm, y_train_mm)
    ridge_ss.fit(X_train_ss, y_train_ss)
    lasso_ss.fit(X_train_ss, y_train_ss)
    ridge_coef_mm.append(ridge_mm.coef_)
    lasso_coef_mm.append(lasso_mm.coef_)
    ridge_coef_ss.append(ridge_ss.coef_)
    lasso_coef_ss.append(lasso_ss.coef_)
    ridge_intercept_mm.append(ridge_mm.intercept_)
    lasso_intercept_mm.append(lasso_mm.intercept_)
    ridge_intercept_ss.append(ridge_ss.intercept_)
    lasso_intercept_ss.append(lasso_ss.intercept_)
```

```python
#For data set which was scaled using Min Max:
for i in range(len(rate_list)):
    print('For alpha=', rate_list[i], ':')
    print('Using Lasso:')
    for j, k in enumerate(X_train_mm.columns):
        print('Column \'', k, '\' has coefficient ', ridge_coef_mm[i][j], sep='')
    print('Intercept is', ridge_intercept_mm[i])
    print('Using Ridge:')
    for j, k in enumerate(X_train_mm.columns):
        print('Column \'', k, '\' has coefficient ', lasso_coef_mm[i][j], sep='')
    print('Intercept is', lasso_intercept_mm[i])
```

```python
#For data set which was scaled using Standard Scaler:
for i in range(len(rate_list)):
    print('For alpha=', rate_list[i], ':')
    print('Using Lasso:')
    for j, k in enumerate(X_train_ss.columns):
        print('Column \'', k, '\' has coefficient ', ridge_coef_ss[i][j], sep='')
    print('Intercept is', ridge_intercept_ss[i])
    print('Using Ridge:')
    for j, k in enumerate(X_train_ss.columns):
        print('Column \'', k, '\' has coefficient ', lasso_coef_ss[i][j], sep='')
    print('Intercept is', lasso_intercept_ss[i])
```

```python
import statsmodels.api as sm
```

Checking the Multicollinearity for Standard Scaled Data

```python
X = df_ss_scaled[df.columns.drop('Chance of Admit ')]
Y = df_ss_scaled["Chance of Admit "]
X_sm = sm.add_constant(X)
sm_model = sm.OLS(Y, X_sm).fit()
print(sm_model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X_t = X
vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### 2) The mean of residuals is nearly zero

```python
print('The mean of residuals for Linear Regression using Min Max Scaled data is',np.mean(l_mm.predict(x_mm_scaled) - y_mm_scaled))
print('The mean of residuals for Linear Regression using Standard Scaled data is',np.mean(l_ss.predict(x_ss_scaled) - y_ss_scaled))
```

```python
#H0: Means for the residuals for Min Max Scaled data is 0
from scipy.stats import ttest_1samp
alpha = 0.05
stat, p_value = ttest_1samp(l_mm.predict(x_mm_scaled) - y_mm_scaled, 0)
if p_value < alpha:
    print("NULL hypothesis is rejected. Mean of residuals is not close to 0.")
else:
    print("NULL hypothesis is not rejected. Mean of residuals is close to 0.")
    print('p-value=', p_value)
```

```python
#H0: Means for the residuals for Standard Scaled data is 0
alpha = 0.05
stat, p_value = ttest_1samp(l_ss.predict(x_ss_scaled) - y_ss_scaled, 0)
if p_value < alpha:
    print("NULL hypothesis is rejected. Mean of residuals is not close to 0.")
else:
    print("NULL hypothesis is not rejected. Mean of residuals is close to 0.")
    print('p-value=', p_value)
```

## 5. Model performance evaluation

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
def adj_r2(r2, x):
    m = len(x)
    d = x.shape[1]
    return 1 - (1 - r2) * (m - 1) / (m - 1 - d)
```

Performance for Min Max Scaled data:

```python
print('For Training data')
print('MAE =', mean_absolute_error(y_train_mm, l_mm.predict(X_train_mm)))
print('RMSE =', mean_squared_error(y_train_mm, l_mm.predict(X_train_mm), squared=0))
print('R2 =', l_mm.score(X_train_mm, y_train_mm))
print('Adj R2 =', adj_r2(l_mm.score(X_train_mm, y_train_mm), X_train_mm))
```

```python
print('For Test data')
print('MAE =', mean_absolute_error(y_test_mm, l_mm.predict(X_test_mm)))
print('RMSE =', mean_squared_error(y_test_mm, l_mm.predict(X_test_mm), squared=0))
print('R2 =', l_mm.score(X_test_mm, y_test_mm))
print('Adj R2 =', adj_r2(l_mm.score(X_test_mm, y_test_mm), X_test_mm))
```

```python
print('For all data (both test and train)')
print('MAE =', mean_absolute_error(y_mm_scaled, l_mm.predict(x_mm_scaled)))
print('RMSE =', mean_squared_error(y_mm_scaled, l_mm.predict(x_mm_scaled), squared=0))
print('R2 =', l_mm.score(x_mm_scaled, y_mm_scaled))
print('Adj R2 =', adj_r2(l_mm.score(x_mm_scaled, y_mm_scaled), x_mm_scaled))
```

Performance for Standard Scaled data:

```python
print('For Training data')
print('MAE =', mean_absolute_error(y_train_ss, l_ss.predict(X_train_ss)))
print('RMSE =', mean_squared_error(y_train_ss, l_ss.predict(X_train_ss), squared=0))
print('R2 =', l_ss.score(X_train_ss, y_train_ss))
print('Adj R2 =', adj_r2(l_ss.score(X_train_ss, y_train_ss), X_train_ss))
```

```python
print('For Test data')
print('MAE =', mean_absolute_error(y_test_ss, l_ss.predict(X_test_ss)))
print('RMSE =', mean_squared_error(y_test_ss, l_ss.predict(X_test_ss), squared=0))
print('R2 =', l_ss.score(X_test_ss, y_test_ss))
print('Adj R2 =', adj_r2(l_ss.score(X_test_ss, y_test_ss), X_test_ss))
```

```python
print('For all data (both test and train)')
print('MAE =', mean_absolute_error(y_ss_scaled, l_ss.predict(x_ss_scaled)))
print('RMSE =', mean_squared_error(y_ss_scaled, l_ss.predict(x_ss_scaled), squared=0))
print('R2 =', l_ss.score(x_ss_scaled, y_ss_scaled))
print('Adj R2 =', adj_r2(l_ss.score(x_ss_scaled, y_ss_scaled), x_ss_scaled))
```

1. MAE and RMSE are more for Standard scaled data compared to MinMax scaled data.
2. R2 and adjusted R2 score are somewhat same for both the scaled data.
3. As expected training R2 and adjusted R2 score are more than that for test data.
4. We can improve the total data (rows) so as to improve the performance of the model.
